```python
# library installation
import requests

URL = "https://timesofmalta.com/news/business"
r = requests.get(URL)

print(r.text)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.content, 'html5lib')
print(soup.prettify())
```

```python
import json

articles = []

table = soup.find('script', attrs = {'type': 'application/ld+json'})

print(table)
```

```python
for row in table:
    data = json.loads(row.string)

    for article in data.get('@graph', []):
        if article.get('@type') == 'NewsArticle':
            
            headline = article.get('headline')
            url = article.get('url')
            articles.append({'headline': headline, 'url': url})

print(articles)
```

```python
import pandas as pd

filename = 'timesofmaltaArticles.csv'

df = pd.DataFrame(articles)

df.to_csv(fr'C:\Users\ivane\Dissertation\data\{filename}', index=False)
```
