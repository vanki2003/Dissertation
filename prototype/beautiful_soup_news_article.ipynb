```python
articles = []

filename = 'maltabusinessweeklyArticles.csv'

page_number = 1
max_pages = 90

while page_number <= max_pages:
    URL = 'http://maltabusinessweekly.com/category/malta-economy/page/' + str(page_number) + '/'
    new_page = requests.get(URL)
    soup = BeautifulSoup(new_page.text, 'html')
    business_article = soup.find_all('a', class_ = "td-image-wrap")
    
    for article in business_article:
        title = article.get('title')
        url = article.get('href')
        articles.append({'headline': title, 'url': url})
        
    dataset = pd.DataFrame(articles)
    dataset.to_csv(fr'C:\Users\ivane\Desktop\Dissertation\data\{filename}', index=False)    
    page_number += 1
    
existing_data = pd.read_csv(fr'C:\Users\ivane\Desktop\Dissertation\data\{filename}')
existing_data.drop_duplicates(subset=['url'], keep='first', inplace=True)
existing_data.to_csv(fr'C:\Users\ivane\Desktop\Dissertation\data\{filename}', index=False)
```

```python
articles = []

filename = 'maltabusinessweeklyArticles.csv'

if 'description' not in existing_data.columns:
    existing_data['description'] = None
if 'date' not in existing_data.columns:
    existing_data['date'] = None

for index, row in existing_data.iterrows():
    try:
        URL = str(row['url'])
        page = requests.get(URL)
        soup = BeautifulSoup(page.text, 'html.parser')
        
        meta_info = soup.find_all('meta')
        description = None
        date = None
        
        for meta in meta_info:
            if meta.get('property') == 'og:description':
                description = meta.get('content')
            if meta.get('property') == 'article:published_time':
                date = meta.get('content')
        
        existing_data.at[index, 'description'] = description
        existing_data.at[index, 'date'] = date
    
    except Exception as e:
        print(f"Error fetching {URL}: {e}")

existing_data.to_csv(fr'C:\Users\ivane\Desktop\Dissertation\data\{filename}', index=False)
        
    
```
