```python
articles = []

filename = 'politicoArticles.csv'

page_number = 1
max_pages = 122

while page_number <= max_pages:
    URL = 'https://www.politico.eu/section/economics/page/' + str(page_number)
    new_page = requests.get(URL)
    soup = BeautifulSoup(new_page.text, 'html')
    business_article = soup.find_all('div', class_='card__content')
    
    for article in business_article:
        title = article.find('h2', class_='card__title').get_text(strip=True) if article.find('h2', class_='card__title') else None
        date = article.find('span', class_='date-time__date').get_text(strip=True) if article.find('span', class_='date-time__date') else None
        time = article.find('span', class_='date-time__time').get_text(strip=True) if article.find('span', class_='date-time__time') else None
        description = article.find('div', class_='card__excerpt').get_text(strip=True) if article.find('div', class_='card__excerpt') else None
        articles.append({'headline': title, 'date': date, 'time': time, 'description': description})
        
    dataset = pd.DataFrame(articles)
    dataset.to_csv(fr'C:\Users\ivane\Desktop\Dissertation\data\{filename}', index=False)    
    page_number += 1
```

```python
articles = []

filename = 'politicoArticles.csv'

page_number = 1
max_pages = 127

while page_number <= max_pages:
    URL = 'https://www.politico.eu/section/economics/' + str(page_number)
    new_page = requests.get(URL)
    soup = BeautifulSoup(new_page.text, 'html')
    business_article = soup.find_all('article')
    
    for article in business_article:
        title = article.find('h4').get_text(strip=True) if article.find('h4') else None
        date = article.find('span').get_text(strip=True) if article.find('span') else None
        description = article.find('p').get_text(strip=True) if article.find('p') else None
        articles.append({'headline': title, 'date': date, 'description': description})
        
    dataset = pd.DataFrame(articles)
    dataset.to_csv(fr'C:\Users\ivane\Desktop\Dissertation\data\{filename}', index=False)    
    page_number += 1
```
